```python
import matplotlib as plt
import numpy as np
x = np.linspace(0,5,10)
y = x ** 2
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')                                 


plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#  we can do this by using objects also
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)

axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2.plot(y,x)
axes2.set_title('Smaller plot')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1 , figsize=(8,5))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.plot(y,x)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x **2)
ax.plot(x,x** 3)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x **2, label= 'X squared')
ax.plot(x,x** 3,label = 'x cubed')
ax.legend()
```
