```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
'A': ['A0', 'A1', 'A2', 'A3'],
'B': ['BO', 'B1', 'B2', 'B3'],
'C': ['co', 'C1', 'C2', 'C3'],
'D': ['DO', 'D1', 'D2', 'D3']},
index=[0, 1, 2, 3])
df1
```

```python
df2 = pd.DataFrame({
'A': ['A4', 'A5', 'A6', 'A7'],
'B': ['B4', 'B5', 'B6', 'B7'],
'C': ['C4', 'C5', 'C6', 'C7'],
'D': ['D4', 'D5', 'D6', 'D7']},
index = [4, 5, 6, 7])
df2
```

```python
df3 = pd.DataFrame({
'A': ['A8', 'A9', 'A10', 'A11'],
'B': ['B8', 'B9', 'B10', 'B11'],
'C': ['C8', 'C9', 'C10', 'C11'],
'D': ['D8', 'D9', 'D10', 'D11']},
index = [8, 9, 10, 11])
df3
```

```python
# concatenation basically glues together the  dataframe  
pd.concat([df1,df2,df3])                            #by the rows
```

```python
pd.concat([df1,df2,df3] , axis =1)                #by the coloums
```

```python
# Merging
# The merge function allows you to merge DataFrames together using a similar logic as merging SQL Tables together. For example:
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
'A': ['A', 'A1', 'A2', 'A3'],
'B': ['BO', 'B1', 'B2', 'B3']})
left
```

```python
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
'C': ['co', 'C1', 'C2', 'C3'],
'D': ['DO', 'D1', 'D2', 'D3']})
right
```

```python
pd.merge(left,right  , on='key')            #on is used to used to the by which coloum we have to merge : 
```

```python
# one more example :: 

left = pd.DataFrame({'key1': ['KO', 'KO', 'K1', 'K2'],
'key2': ['K0', 'K1', 'KO', 'K1'],
'A': ['A0', 'A1', 'A2', 'A3'],
'B': ['BO', 'B1', 'B2', 'B3']})


right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
'key2': ['K0', 'к0', 'KO', 'к0'],
'C': ['co', 'C1', 'C2', 'C3'],
'D': ['DO', 'D1', 'D2', 'D3']})


# merging the two datfames by two coloums :: 
pd.merge(left, right,how="outer", on=['key1', 'key2'])        

#how is used to specifiy the type of merging you want , learn this is sql 
```

```python
# join
# join is method used to combine the two coloum of differently indexed dataframe
# first dataframe::
x = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
'B': ['BO', 'B1', 'B2']},
index=['K0', 'K1', 'K2'])

# second dataframe::
y = pd.DataFrame({'C': ['CO', 'C2', 'C3'],
'D': ['DO', 'D2', 'D3']},
index=['K0', 'K2', 'K3'])

x
```

```python
y
```

```python
x.join(y)
```

```python
x.join(y ,how='outer')
```
