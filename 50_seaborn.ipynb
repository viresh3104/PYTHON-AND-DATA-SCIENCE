```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head(5)
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.jointplot(x='total_bill' , y = 'tip' ,data=tips )                        #x and y are name of coloums 
```

```python
sns.jointplot(x='total_bill' , y = 'tip' ,data=tips , kind = 'hex')         #kind parameter helps us to customize the scatter plot
# sns.jointplot(x='total_bill' , y = 'tip' ,data=tips , kind = 'reg')
# sns.jointplot(x='total_bill' , y = 'tip' ,data=tips , kind = 'kde')
```

```python
sns.pairplot(tips)    
# like to joinplot pair plot give relation between all possible combination of dataframe
# but in joinplot we have to specify the relation ship bet two coloumns 
```

```python
sns.pairplot(tips , hue='sex')           #hue parameter classify the categories in all combinations    
```

```python
sns.rugplot(tips['total_bill'])                   #it simply draw one line for each point ,unline to the histogram
```

```python
sns.kdeplot(tips['total_bill'])
```
