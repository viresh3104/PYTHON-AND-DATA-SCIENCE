```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
%matplotlib inline

df = pd.read_csv('Sales Data.csv')
df = df.head(1000)
df
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Day of Week'] = df['Order Date'].dt.day_name()
df['Month'] = df['Order Date'].dt.month
df
```

```python
# Histogram of Sales Amounts: 
# Plot a histogram to visualize the distribution of sales amounts.

sns.histplot(df['Sales'] , bins= 30 , kde=True)
plt.title('Distribution of Sales Amounts')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
```

```python
# Pie Chart of Orders by Hour: 
# Create a pie chart showing the proportion of orders placed in each hour of the day. 
plt.figure(figsize=(10,6))
order_counts = df['Hour'].value_counts().sort_index()
order_counts.plot(kind='pie' , autopct='%1.1f%%' , startangle = 140)
plt.title('Proportion of Orders Placed Each Hour')
```

```python
# Bar Plot of Average Price Each by Product: 
# Plot a bar chart showing the average price of each product. 

avg = df.groupby('Product')['Price Each'].mean().sort_values()
plt.figure(figsize=(10,6))
sns.barplot(x= avg.index , y = avg.values ,  palette='viridis')
plt.title('Average Price of Each Product')
plt.xlabel('Product')
plt.ylabel('Average Price')
plt.xticks(rotation=90)
```

```python
# Scatter Plot of Sales vs Quantity Ordered: 
# Create a scatter plot to examine the relationship between sales amount and quantity ordered.

sns.scatterplot(x=df['Sales'] , y =df["Quantity Ordered"])
plt.title('Sales vs Quantity Ordered')
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales')
```

```python
# Stacked Bar Plot of Sales by Product and City: 
# Create a stacked bar plot showing total sales for each product, stacked by city.

kd = df.groupby(['Product','City'])['Sales'].sum().unstack()
kd.plot(kind='bar' , stacked=True , colormap='viridis')
plt.title('Total Sales by Product and City')
plt.xlabel('Product')
plt.ylabel('Total Sales')
```

```python
# Horizontal Bar Plot of Average Quantity Ordered by City: 
# Plot a horizontal bar chart showing the average quantity ordered from each city.

hg = df.groupby('City')['Quantity Ordered'].mean().reset_index()
sns.barplot(x = 'Quantity Ordered' , y = 'City' , data=hg)
plt.title('Average Quantity Ordered by City')
plt.xlabel('Average Quantity Ordered')
plt.ylabel('City')
```

```python
# Area Plot of Quantity Ordered Over Time: 
# Create an area plot to visualize the trend of quantity ordered over time

dfp = pd.read_csv('Sales Data.csv')
quantity_over_time = dfp.groupby('Month')['Quantity Ordered'].count()
quantity_over_time.plot(kind='area', colormap='viridis')
plt.title('Quantity Ordered Over Time')
plt.xlabel('Time')
plt.ylabel('Quantity Ordered')
```

```python
# Bar Plot of Sales by Day of the Week: 
# Create a bar chart showing total sales for each day of the week.

group = df.groupby('Day of Week')['Sales'].sum().reset_index()
sns.barplot(x = 'Day of Week' ,  y = 'Sales' , data= group)
```

```python
# Multi-line Plot of Quantity Ordered by Product Over Hours: 
# Create a multi-line plot to show the trend of quantity ordered for each product over the hours of the day. 

group = df.groupby(['Hour'  , 'Product'])['Quantity Ordered'].sum().unstack()
plt.figure(figsize=(15,10))
sns.lineplot(data=group)
plt.title('Quantity Ordered by Product Over Hours')
plt.xlabel('Hour of the Day')
plt.ylabel('Quantity Ordered')
plt.legend(title='Product')
```

```python
# Box Plot of Sales by Hour: 
# Create a box plot to visualize the distribution of sales amounts for each hour of the day. 
plt.figure(figsize=(10,8))
sns.boxplot(x= df['Sales'] , y = df['Hour'])
plt.title('Box Plot of Sales by Hour')
plt.xlabel('Sales Amount')
plt.ylabel('Hour of the Day')
plt.xticks(rotation = 90)
plt.yticks(range(0,24))
plt.grid()
```

```python
# Point Plot of Quantity Ordered by Month and Product: 
# Create a point plot to show the trend of quantity ordered over the months, differentiated by product. 
df = pd.read_csv('Sales Data.csv')
plt.figure(figsize=(12,8))
sns.pointplot(x = 'Month' , y = 'Quantity Ordered' ,  hue = 'Product' ,data = df )
plt.title('Quantity Ordered by Month and Product')
plt.xlabel('Month')
plt.ylabel('Quantity Ordered')
plt.legend(title='Product')
```

```python
# Count Plot of Orders by Day of the Week: 
# Use a count plot to show the number of orders for each day of the week. 
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Day of Week'] = df['Order Date'].dt.day_name()
plt.figure(figsize=(12,8))

sns.countplot(x='Day of Week', data=df, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], palette='viridis')

plt.title('Number of Orders by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Orders')
```

```python
# Heatmap of Sales by Product and Hour: 
# Create a heatmap to visualize sales data across products and hours. 

sales_by_product_hour = df.groupby(['Product', 'Hour'])['Sales'].sum().unstack()
plt.figure(figsize=(12,8))

sns.heatmap(sales_by_product_hour, fmt=".1f", cmap="viridis")

plt.title('Heatmap of Sales by Product and Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Product')
```

```python
# Bar Plot of Average Sales by City: 
# Use a bar plot to show the average sales amount for each city. 

avg_sales_per_city = df.groupby('City')['Sales'].mean().sort_values()
sns.barplot(x=avg_sales_per_city.values, y=avg_sales_per_city.index, palette='viridis')
plt.title('Average Sales by City')
plt.xlabel('Average Sales')
plt.ylabel('City')
```

```python
# Violin Plot of Quantity Ordered by Product: 
# Draw a violin plot to display the distribution of quantities ordered for each product. 
sns.violinplot(x='Quantity Ordered', y='Product', data=df, palette='viridis')
plt.title('Distribution of Quantities Ordered by Product')
plt.xlabel('Product')
plt.ylabel('Quantity Ordered')
```

```python
# Facet Grid of Sales by Month and Product: 
# Use a Facet Grid to plot sales data by month, separated by product. 
g = sns.FacetGrid(df, col='Product', col_wrap=4, height=4, palette='viridis')
g.map(sns.lineplot, 'Month', 'Sales')
g.add_legend()
plt.subplots_adjust(top=0.9)
g.fig.suptitle('Sales by Month and Product')
```

```python
# Pair Plot of Quantity Ordered and Sales by Product: 
# Create a pair plot to explore the relationship between quantity ordered and sales for each product. 

sns.pairplot(df, hue='Product', vars=['Quantity Ordered', 'Sales'], palette='viridis')
```

```python
# Box Plot of Sales by Month: 
# Create a box plot to visualize the distribution of sales amounts for each month. 

plt.figure(figsize=(12,8))
sns.boxplot(x='Sales', y='Month', data=df)
plt.title('Sales by Month')
plt.xlabel('Sales')
plt.xticks(rotation = 90 )
plt.ylabel('Month')
```
