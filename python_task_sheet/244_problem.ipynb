```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

df = pd.read_csv('Sales Data.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['day'] = df['Order Date'].dt.day
df
```

```python
df = pd.DataFrame(df)
df
```

```python
# Line Plot of Cumulative Sales Over Time: 
# Plot a line chart showing the cumulative sales over time

x  = df['Sales'].cumsum()
sns.lineplot(x=df['Month'] , y = x)
plt.grid()
```

```python
# Bar Plot of Number of Orders by Month: 
# Create a bar chart to show the number of orders for each month

x = df.groupby('Month')['Order ID'].count().reset_index()

sns.barplot(x= 'Month', y = 'Order ID'  , data=x)
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.title('Number of Orders by Month')
```

```python
# Pie Chart of Quantity Ordered by Product: 
# Create a pie chart to show the proportion of the total quantity ordered for each product

x = df.groupby('Product')['Quantity Ordered'].sum()
plt.figure(figsize=(8,6))
x.plot(kind='pie' , startangle=130 , autopct='%1.1f%%')

plt.title('Total Sales by City')
```

```python
# Scatter Plot of Sales by Order Date: 
# Plot a scatter plot to show sales amounts against order dates.

day_sale = df.groupby('day')['Sales'].sum().reset_index()
day_sale

sns.scatterplot(x = 'day' , y ='Sales', data=day_sale)
plt.xticks(range(1,31))
plt.tight_layout()
plt.grid()
```

```python
# Stacked Bar Plot of Sales by Month and City: 
# Create a stacked bar plot to show total sales for each month, stacked by city.

group = df.groupby(['Month' ,'City'])['Sales'].sum().unstack()
group.plot(kind='bar' , stacked=True , figsize=(10,8))
plt.xlabel ('Month')
plt.ylabel('Total Sales')
```

```python
# Horizontal Bar Plot of Total Sales by City: 
# Plot a horizontal bar chart showing total sales for each city
group = df.groupby('City')['Sales'].sum().reset_index()

sns.barplot(x='Sales' , y = 'City' , data=group)
```

```python
# Area Plot of Sales Over Time: 
# Create an area plot to visualize sales trends over time.
av = df.groupby('Month')['Sales'].sum()
av.plot(kind='area')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(range(1,13))
plt.title('Total Sales In Each Month')
```

```python
# Line Plot of Average Sales by Hour:
# Plot a line chart showing the average sales amount for each hour of the day

group = df.groupby('Hour')['Sales'].mean().reset_index()
sns.lineplot(x = 'Hour' , y= 'Sales' , data=group)
```

```python
# Bubble Plot of Price vs Quantity (Size by Sales):
# Create a bubble plot where each point's size represents the total sales amount, showing the
#  relationship between price and quantity ordered.


aw = df.groupby('Price Each')['Quantity Ordered'].sum().reset_index()

plt.figure(figsize=(15,10))
sns.scatterplot(data=aw , x ='Price Each' , y='Quantity Ordered' , size='Quantity Ordered' ,sizes=(100,900) , hue=df['Product'],palette="rainbow")

```

```python
# Multi-line Plot of Sales by Product Over Months:
# Create a multi-line plot to show the sales trend of each product over the months.

group = df.groupby(['Product' , 'Month' ])['Sales'].sum().unstack()


group.plot(kind="line" , figsize=(20,10))
```

```python
# Line Plot of Average Sales by Month:
# Use a line plot to show the average sales amount for each month.

x = df.groupby('Month')['Sales'].mean()
sns.lineplot(data=x )
plt.xticks(range(1,13))
plt.grid()
```

```python
# Count Plot of Orders by Product:
# Create a count plot to display the number of orders for each product.
# x = df.groupby('Product')['Order ID'].count().reset_index()

sns.countplot(x = 'Product' , data=df)



plt.xlabel('Product')
plt.ylabel('Number of Orders')
plt.title('Number of Orders by Product')
plt.xticks(rotation=90)
```

```python
# Heatmap of Sales by City and Month: 
# Create a heatmap to visualize sales data across cities and months

pt = df.pivot_table(index='City' , columns='Month' , values='Sales')
plt.figure(figsize=(10,8))
sns.heatmap(pt )
```

```python
# Bar Plot of Total Quantity Ordered by City: 
# Use a bar plot to show the total quantity ordered from each city.

xs = df.groupby('City')['Quantity Ordered'].sum().reset_index()

sns.barplot(x= 'City' , y = 'Quantity Ordered' , data=xs )
plt.xticks(rotation = 90)
```

```python
# Violin Plot of Sales by Product: 
# Draw a violin plot to display the distribution of sales amounts for each product.

sns.violinplot(data= df , x='Product' , y= "Sales")

plt.xlabel('Product')
plt.ylabel('Sales Amount')
plt.title('Distribution of Sales Amounts by Product')
plt.xticks(rotation = 90)
```

```python
# Facet Grid of Quantity Ordered by Hour and Product: 
# Use a Facet Grid to plot quantity ordered by hour, separated by product.

xs = sns.FacetGrid(data=df , col='Product' , col_wrap=4)
xs.map(sns.lineplot ,"Hour" , 'Quantity Ordered')
xs.figure.suptitle('Product and Quantity Ordered by Hour',y=1.05) 
```

```python
# Line Plot of Total Sales by City Over Time: 
# Create a line plot to show total sales for each city over time

xs = df.groupby(['Month','City'])['Sales'].sum().reset_index()
plt.figure(figsize=(20,10))
sns.lineplot(data=xs ,x = 'Month', y='Sales', hue = 'City')
plt.xticks(range(1,13))
plt.grid()
```

```python
# Strip Plot of Quantity Ordered by Month: 
# Use a strip plot to display individual quantity ordered across different months.

xs = df.groupby('Month')['Quantity Ordered'].sum().reset_index()
plt.figure(figsize=(10,10))
sns.stripplot(data = xs , x = 'Month' , y= 'Quantity Ordered')
```

```python
# Boxen Plot of Sales by Hour: 
# Create a boxen plot to visualize the distribution of sales amounts for each hour of the day.
plt.figure(figsize=(20,10))
sns.boxplot(data = df ,  x ='Sales' , y = 'Hour')
plt.xticks(rotation = 90)
plt.yticks(range(1,25))

```

```python
# Swarm Plot of Quantity Ordered by Product: 
# Use a swarm plot to show the distribution of quantity ordered for each product.

sns.swarmplot(x = 'Quantity Ordered' , y = 'Product' , data = df.head(1000))
plt.xticks(rotation = 90)
```
