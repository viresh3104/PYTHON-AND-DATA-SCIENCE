```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('Sales Data.csv')
df = pd.DataFrame(df)
df
```

```python
# Bar Plot of Quantity Ordered by Product: 
# Plot a bar chart showing the total quantity ordered for each product. 

total_quantity = df.groupby('Product')['Quantity Ordered'].sum().sort_values()
plt.figure(figsize=(10,8))
total_quantity.plot(kind='barh')
plt.xlabel('Quantity Ordered')
plt.ylabel('Product')
```

```python
# Line Plot of Sales Over Time: 
# Create a line plot to show how sales have varied over the months

total_sales_over_time = df.groupby('Month')['Sales'].sum()
plt.figure(figsize=(8,6))
total_sales_over_time.plot(kind='line' )
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(range(1, 13))

plt.grid(True)
```

```python
# Histogram of Order Quantity: 
# Draw a histogram to visualize the distribution of the quantities ordered

plt.figure(figsize=(8,6))
plt.hist(df['Quantity Ordered'] , bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Quantities Ordered')
plt.xlabel('Quantity Ordered')
plt.ylabel('Frequency')
```

```python
# Pie Chart of Sales by City: 
# Create a pie chart to show the proportion of total sales coming from each city

group_by_city = df.groupby('City')['Sales'].sum()
plt.figure(figsize=(8,6))
group_by_city.plot(kind="pie",startangle=160 , autopct='%1.1f%%')
plt.title('Total Sales by City')
```

```python
# Scatter Plot of Price vs Quantity:
# Plot a scatter plot to examine the relationship between the price of a product and the quantity ordered

# Plotting the scatter plot
# z= df.groupby('Product')['Quantity Ordered'].count()

plt.figure(figsize=(10, 6))

plt.scatter(x=df['Price Each'] ,y=df['Quantity Ordered'])
plt.title('Price vs Quantity Ordered')
plt.xlabel('Price Each')
plt.ylabel('Quantity Ordered')
```

```python
# Bar Plot of Total Sales by Month: 
# Plot a bar chart to show total sales for each month

group = df.groupby('Month')['Sales'].sum()

plt.figure(figsize=(10,8))
group.plot(kind='bar')

plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
```

```python
# Line Plot of Sales Over Hours: 
# Create a line plot to show how sales change throughout the day (by hour).

group = df.groupby('Hour')['Sales'].sum()

plt.figure(figsize=(10,8))
group.plot(kind='line')
plt.title('Sales Over Hours')
plt.xlabel('Hour')
plt.ylabel('Total Sales')
plt.xticks(range(24))
plt.grid()
```

```python
# Horizontal Bar Plot of Sales by Product: 
# Plot a horizontal bar chart showing total sales for each product


total_quantity = df.groupby('Product')['Sales'].sum().sort_values()
plt.figure(figsize=(10,8))
total_quantity.plot(kind='barh')
plt.xlabel('Sales')
plt.ylabel('Product')
```

```python
# Stacked Bar Plot of Quantity Ordered by Product and Month: 
# Create a stacked bar plot showing the quantity ordered for each product, stacked by month.

product_month_quantity = df.groupby(['Month', 'Product'])['Quantity Ordered'].sum().unstack()

product_month_quantity.plot(kind='bar', stacked=True, figsize=(14, 14))
plt.title('Quantity Ordered by Product and Month')
plt.xlabel('Month')
plt.ylabel('Quantity Ordered')
# plt.legend(title='Product')
```

```python
# Bar Plot of Number of Orders by Hour: 
# Plot a bar chart showing the number of orders placed in each hour of the day.

hourly_orders = df.groupby('Hour')['Order ID'].count()


plt.figure(figsize=(10, 6))
hourly_orders.plot(kind='bar', color='skyblue')
plt.title('Number of Orders by Hour')
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.xticks(range(24))
plt.grid(True)
```
