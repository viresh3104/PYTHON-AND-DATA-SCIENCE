```python
import pandas as pd
df = pd.read_csv('Sales Data.csv')
df
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df
```

```python
df['hour'] = df['Order Date'].dt.hour
df
```

```python
df['year'] = df['Order Date'].dt.year
df
```

```python
df['day'] = df['Order Date'].dt.day
df
```

```python
# How to find the top 5 products by total sales?
top_5_product = df.groupby('Product')['Sales'].sum()
top_5_product.sort_values(ascending=False).head(5)
```

```python
# How to find the top 5 cities by total sales?
top_5_cities = df.groupby('City')['Sales'].sum()
top_5_cities.sort_values(ascending=False).head(5)
```

```python
# How to find the top 5 hours by total sales?
top_5_hours = df.groupby('hour')['Sales'].sum()
top_5_hours.sort_values(ascending=False).head(5)
```

```python
# How to find the bottom 5 products by total sales?
bottom_5_product = df.groupby('Product')['Sales'].sum()
bottom_5_product.sort_values().head(5)
```

```python
# How to find the bottom 5 cities by total sales?
bottom_5_cities = df.groupby('City')['Sales'].sum()
bottom_5_cities.sort_values().head(5)
```

```python
# How to find the bottom 5 hours by total sales?
bottom_5_hours = df.groupby('hour')['Sales'].sum()
bottom_5_hours.sort_values().head(5)
```

```python
# How to find the most frequently ordered product?
most_frequently_orderd_product = df.groupby(['Product'])['Quantity Ordered'].sum()
most_frequently_orderd_product.sort_values(ascending=False).head(5)
```

```python
# How to find the least frequently ordered product?
least_frequently_orderd_product = df.groupby(['Product'])['Quantity Ordered'].sum()
least_frequently_orderd_product.sort_values().head(5)
```

```python
# How to find the average order size (total sales divided by number of orders) for each city?
avg_order_size = df.groupby('City')['Sales'].mean()
avg_order_size.sort_values(ascending=False).head(5)
```

```python
# How to calculate the percentage contribution of each product to total sales?
total_sales_per_product = df.groupby('Product')['Sales'].sum()
total_sales_of_all_product = total_sales_per_product.sum()



perc_contribution_product = ( total_sales_per_product/ total_sales_of_all_product)*100
perc_contribution_product
```

```python
# How to calculate the percentage contribution of each city to total sales?
total_sales_per_city = df.groupby('City')['Sales'].sum()
total_sales = total_sales_per_city.sum()

perc_contribution_city = (total_sales_per_city / total_sales) *100
perc_contribution_city
```

```python
# How to calculate the percentage contribution of each month to total sales?
total_sales_per_month = df.groupby('Month')['Sales'].sum()
total_sales_of_all_monts = total_sales_per_month.sum()

perc_contribution_month = (total_sales_per_month / total_sales_of_all_monts) *100
perc_contribution_month
```

```python
# How to calculate the percentage contribution of each hour to total sales?
total_sales_per_hour = df.groupby('hour')['Sales'].sum()
total_sales_of_all_hours = total_sales_per_hour.sum()

perc_contribution_hour = (total_sales_per_hour / total_sales_of_all_hours) *100
perc_contribution_hour
```

```python
# How to find the most common order size (quantity ordered)?
most_common = df['Quantity Ordered'].value_counts()
most_common = most_common.idxmax()
most_common
```

```python
# How to find the most common order value (sales)?
most_common_order_value = df['Sales'].value_counts()
most_common_order_value = most_common_order_value.idxmax()
most_common_order_value
```

```python
# How to calculate the cumulative sales over time?
cumulative_sales = df['Sales']
cumulative_sales.cumsum()
```

```python
# How to find the days with the highest total sales?
highest_sales = df.groupby('day')['Sales'].sum()
highest_sales.sort_values(ascending=False).head(10)
```

```python
# How to find the days with the lowest total sales?
lowest_sales = df.groupby('day')['Sales'].sum()
lowest_sales.sort_values().head(10)
```

```python
# How to find the products that generated the most revenue in each city?
city_product_sales = df.groupby(['City', 'Product'])['Sales'].sum()
# most_revenue_product_per_city = city_product_sales.loc[city_product_sales.groupby('City')['Sales'].idxmax()]
```
