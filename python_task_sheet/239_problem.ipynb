```python
import pandas as pd
df = pd.read_csv('Sales Data.csv')
df = df.head(1000)
df
```

```python
# How to convert a column to a different data type (e.g., from string to datetime)
df['Order Date'] = pd.to_datetime(df['Order Date'])
df
```

```python
# How to extract the year from a datetime column?
df['year'] = df['Order Date'].dt.year
df
```

```python
# How to extract the month from a datetime column?
df['month'] = df['Order Date'].dt.month
df
```

```python
# How to extract the day from a datetime column?
df['day'] = df['Order Date'].dt.day
df
```

```python
# How to extract the hour from a datetime column?
df['hour'] = df['Order Date'].dt.hour
df
```

```python
# How to extract the minute from a datetime column?
df['minute'] = df['Order Date'].dt.minute
df
```

```python
# How to extract the second from a datetime column?
df['second'] = df['Order Date'].dt.second
df
```

```python
# How to calculate the total sales for each month?
sales_per_month = df.groupby('Month')['Sales'].sum()
sales_per_month

```

```python
# How to calculate the total sales for each city?
sales_city = df.groupby('City')['Sales'].sum()
sales_city
```

```python
# How to calculate the total sales for each hour of the day?
sales_per_hour_in_day = df.groupby(['day','hour'])['Sales'].sum()
sales_per_hour_in_day
```

```python
# How to calculate the average quantity ordered for each product?
avg_sales_per_product = df.groupby('Product')['Sales'].mean()
avg_sales_per_product
```

```python
# How to calculate the total quantity ordered for each product?
total_quantity_orderd = df.groupby('Product')['Quantity Ordered'].sum()
total_quantity_orderd
```

```python
# How to find the product that has the highest price each?
max_price_for_each_product = df.groupby('Product')['Price Each'].max()
max_price_for_each_product

```

```python
# How to find the product that has the lowest price each?
min_price_for_each_product = df.groupby('Product')['Price Each'].min()
min_price_for_each_product
```

```python
# How to calculate the average price for each product?
avg_price_for_each_product = df.groupby('Product')['Sales'].mean()
avg_price_for_each_product
```

```python
# How to calculate the total number of orders for each month?
total_number_of_orders_per_month = df.groupby(['Month','Product'])['Quantity Ordered'].sum()
total_number_of_orders_per_month
```

```python
# How to calculate the average price for each city?
avg_price_for_each_city = df.groupby('City')['Sales'].mean()
avg_price_for_each_city
```
