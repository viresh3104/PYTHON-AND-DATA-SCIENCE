```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head(5)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
#it works similar to groopy by func
```

```python
sns.countplot(x='sex' ,data=tips) 
# y axis is default set to count in the count plot
```

```python
sns.boxplot(x='day' , y='total_bill' , data=tips)
```

```python
# now in this box plot we split data by spllit per day now we add one more parameter in this which is hue
# lets split further with sex distribution 
sns.boxplot(x='day' , y='total_bill' , data=tips , hue = 'sex')
```

```python
sns.boxplot(x='day' , y='total_bill' , data=tips , hue = 'smoker')
```

```python
sns.violinplot(x='day' , y ='total_bill' ,data= tips)
# box and violine plot are similar to each other but violine plot shows more details 
```

```python
sns.stripplot(x='day' , y ='total_bill' ,data= tips)
# we can use all parametter in this plot also like hue or split or jitter
```

```python
sns.swarmplot(x='day' , y ='total_bill' ,data= tips)
```

```python
# to understand better we can combine the violine and swarmplot
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.catplot(x='day',y='total_bill',data=tips , kind='violin') 
# this is the most efficient way to plot any kind of graph. just change the kind of plot here i put the violine
```
