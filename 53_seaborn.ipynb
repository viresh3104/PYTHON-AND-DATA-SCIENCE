```python
import seaborn as sns
%matplotlib inline
iris= sns.load_dataset('iris')
iris.head()
```

```python
# to checkout the unique values in species coloum
iris['species'].unique()
```

```python
# lets plot a plot with all possible relation which is pairplot
sns.pairplot(iris )    
```

```python
# in the pairplot if we use insted PairGrid then it gives the empty plots
test = sns.PairGrid(iris)
```

```python
test = sns.PairGrid(iris)
test.map_diag(sns.boxenplot)
test.map_upper(sns.kdeplot)
test.map_lower(sns.stripplot)


# we can customize the graphs when we use PlotGrid
```
